```python
import pandas as pd
from pathlib import Path
```

```python
logs_path = Path('logs')
logs_path.exists()

dfs = []
for log in logs_path.glob('CIFAR10_10_public_150*'):
    for csv in log.glob('*.csv'):
        df = pd.read_csv(csv)
        dfs.append(df)
dff = pd.concat(dfs, ignore_index=True)
dff['dataset'] = 'CIFAR10'
dff
        
```
